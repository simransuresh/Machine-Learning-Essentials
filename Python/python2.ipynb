```python
# map - executing the function for every element in the iterable
list1 = [1, 2, 3, 4, 5]

def is_even(num):
  return True if num%2 == 0 else False

print(list(map(is_even, list1)))

```

```python
# filter - filter the elements in the iterable based on a condition
string1 = ["a", "b", "c", "e", "d"]

def filter_vowels(character):
  vowels = ["a", "e", "i", "o", "u"]
  if character in vowels:
    return character
  else: 
    return None

print(list(filter(filter_vowels, string1)))
```

```python
# reduce - reduce the iterable to a single value
import functools

list1 = [1, 2, 3, 4, 5]
print(type(list1)) # type of class

print(functools.reduce(lambda elem1, elem2: elem1+elem2, list1))
```
