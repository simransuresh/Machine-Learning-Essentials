```python
test_list = [1, 4, 5, 7, 7, 8]
# sum = 12 = (4, 8), (5, 7)

# itertools - combinations 
# use list comprehension to get sum
# filter sum = 12
# 
from itertools import combinations
comb_list = list(combinations(test_list, 2))
filt_list = list(set([comb for comb in comb_list if sum(comb) == 12]))
print(comb_list)
print(filt_list)
```

```python
from copy import deepcopy

# call by reference
def function(d1):
  d1["name"] = "new_name"
  print("inside function", d1["name"])

# copy, deepcopy
d1 = {"name": "name1", "age": 10}
# d1 name before manipulation
print(d1["name"])
# send a copy of d1 to manipulate
function(deepcopy(d1))
# d1 is unchanged after manipulation
print(d1["name"])

# global var - login = false


```

```python
def binary_search(arr, low, high, x):
  if high >= low:
 
    mid = (high + low) // 2

    # If element is present at the middle itself
    if arr[mid] == x:
        return mid

    # If element is smaller than mid, then it can only
    # be present in left subarray
    elif arr[mid] > x:
        return binary_search(arr, low, mid - 1, x)

    # Else the element can only be present in right subarray
    else:
        return binary_search(arr, mid + 1, high, x)
 
  else:
      # Element is not present in the array
      return -1
print(binary_search([1, 2, 3, 4, 5], 0, 5, 1))
```
